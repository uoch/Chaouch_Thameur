```python
import os
import googleapiclient.discovery

def get_all_comments(youtube, video_id):
    comments = []

    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=1000  # You can adjust this value based on your needs
    )

    while request:
        response = request.execute()

        for item in response.get("items", []):
            comment = item["snippet"]["topLevelComment"]["snippet"]["textOriginal"]
            comments.append(comment)

            # If there are replies, add them to the comments list
            if "replies" in item:
                for reply_item in item["replies"]["comments"]:
                    reply = reply_item["snippet"]["textOriginal"]
                    comments.append(reply)

        request = youtube.commentThreads().list_next(request, response)

    return comments

def main():
    os.environ["OAUTHLIB_INSECURE_TRANSPORT"] = "1"

    api_service_name = "youtube"
    api_version = "v3"
    DEVELOPER_KEY = "AIzaSyB4gg-044SfIE2ucsUeyNujGL4JVo-uIhk"

    youtube = googleapiclient.discovery.build(
        api_service_name, api_version, developerKey=DEVELOPER_KEY)

    video_id = "2kLyUtVedxo"
    all_comments = get_all_comments(youtube, video_id)

    for i, comment in enumerate(all_comments, start=1):
        print(f"Comment {i}: {comment}")

if __name__ == "__main__":
    main()
```
